# Exploratory Questions
#### Topic: The creation of characters in DC Comics Universe

For this assignment, I want to look at the creation of characters in the history of DC Comics. I want to see whether heroes or villains are better created and presented over the years. For that purpose, I want to ask the following questsions:

```python
dc_chars = pd.read_csv("dc-wikia-data.csv")
dc_chars = dc_chars.drop(columns = ['page_id','urlslug','GSM','EYE','HAIR'])
# clean up the wikia information 
# dc_chars.head()
```

# Discoveries and Insights
<hr> 

## Checking for NaNs
First, I removed all the data that I don't need and only keep the data that I actually use. Then, I want to check the quality of the data by first checking the number of NaNs in the data.

```python
# First, let's look at the distribution of Nans in the dataset
dc_chars.columns = ['name','secret','alignment','sex','alive','appearances','debut month','debut year']
y_pos = np.arange(dc_chars.shape[1])
x_ticks = dc_chars.columns.to_list()
plt.figure(figsize=(15,6))
plt.tight_layout()
plt.bar(y_pos,len(dc_chars) - dc_chars.count(),align='center',alpha = 1, color = 'red', label = 'Number of NaNs')
plt.bar(y_pos,dc_chars.count(),bottom = len(dc_chars) - dc_chars.count(), align='center',alpha = 1, label = 'Number of valid values')
plt.xticks(y_pos,x_ticks)
plt.legend(loc = 'upper center')
dc_chars.rename(columns = {'FIRST APPEARANCE':'MONTH'},inplace=True)
plt.title("Number of missing values in each column of the data", fontsize = 20)
plt.ylabel("Number of Occurances")
plt.xlabel("Variables")
```

From the graph, we can see that there are quite a lot of missing values in the secret identity column of the data. However, there are not that many missing values in the debut month or year. Since I want to focus on the debut month and year, I filter out all entries with no debut month or year values, since there are not that many of those.
<hr>

```python
# clean up the nan data entries
# print(len(dc_chars))
# dc_chars.count()
remove_indexes = []
for index, row in dc_chars.iterrows():
    if pd.isna(row['debut month']) or pd.isna(row['debut year']):
        remove_indexes.append(index)
dc_chars = dc_chars.drop(remove_indexes)
# print(len(dc_chars))
# dc_chars.count()
# change the year and number of appearances to integer
dc_chars['debut year'] = dc_chars['debut year'].fillna(0.0).astype(int)
dc_chars.appearances = dc_chars.appearances.fillna(0.0).astype(int)
# dc_chars.head()
# let's get the month of the first appearence as well, and change it to numerical value
months = ['January','February','March','April','May','June','July','August','September','October','November','December']
for index, row in dc_chars.iterrows():
    if row['debut month']:
        try:
            month = row['debut month'].split(",")[1].strip()
            numerical_month = months.index(str(month)) + 1
            dc_chars.at[index, 'debut month'] = numerical_month
        except:
            dc_chars.at[index, 'debut month'] = 0
# dc_chars.head()
```

## Comparing appearances between heroes and villains
Now, let's compare the number of appearances between the top 20 villains and the top 20 superheroes.

```python
heroes = []
heroes_names = []
villains = []
villains_names = []
for index, row in dc_chars.iterrows():
    if row['alignment'] == 'Good Characters':
        heroes.append(row['appearances'])
        heroes_names.append(row['name'])
    elif row['alignment'] == 'Bad Characters':
        villains.append(row['appearances'])
        villains_names.append(row['name'])
plt.figure(figsize =(15,7))
width = 0.35
plt.bar(np.arange(1,21),heroes[:20],width,label = "Heroes")
plt.bar(np.arange(1,21)+width,villains[:20],width, label = "Villains")
plt.xticks(np.arange(1,21))
plt.title("A comparison of the number of appearances of the top twenty heroes and villains", fontsize = 20)
plt.ylabel("Number of appearances")
plt.xlabel("The top rankings")
plt.legend()
```

So it seems that the top 20 heroes appear way more than the top 20 villains. Here, we can see that the villains are the less-favored ones. This makes sense because comics are mostly about heroes after all, and normally one hero has multiple villains. It is very rare for a villain to have their own comic series.

## Character creation by months
Now, we will look at the character creations by months, Let's first look at the total number of characters created each month


```python
plt.figure(figsize =(15,7))
plt.title("Number of characters debut each month by allignment", fontsize = 20)
months_heroes = [0] * 13
months_villains = [0] * 13
ind = np.arange(13)
width = 0.35
for index, row in dc_chars.iterrows():
    if row['alignment'] == 'Good Characters':
        months_heroes[row['debut month']] += 1
    elif row['alignment'] == 'Bad Characters':
        months_villains[row['debut month']] += 1
plt.bar(ind,months_heroes,width,label = "Hero")
plt.bar(ind+width,months_villains, width,label = "Villain")
plt.xlabel("Month")
plt.ylabel("Number of occurances")
plt.xticks(ind,rotation = 0)
plt.legend()
```

The data with months 0s are data with errors. There are only a small number of them.

Looking at the months aggregated over all the years, I can't see any nothworthy trend here. The overall distribution is roughly uniformed, which is as expected. Even for heroes or villains, the overall distribution is also roughly uniformed. Furthermore, just looking at the monthly data alone, we can see that the number of heroes created is roughly the same as the number of villains created.

## Character created by years
I want to see if there is any trends or patterns in the alignment of the characters created

```python
# Now, let's look at all years that ever have any new character debut
# plt.figure(figsize=(15,7))
# plt.title("Number of character debut each year from 1935 to 2013")
# plt.hist(dc_chars['debut year'],2013-1935)

# Now, let's look at the character debuts based on alignment
# 
print(dc_chars['alignment'].unique())
good_chars = [0] * (2013-1935+1)
bad_chars = [0] * (2013-1935+1)
neutral_chars = [0] * (2013-1935+1)
reformed_criminals = [0] * (2013-1935+1)
x_range= np.arange(1935,2014)
for index, row in dc_chars.iterrows():
    if row['alignment'] == 'Good Characters':
        good_chars[row['debut year']-1935] += 1
    elif row['alignment'] == 'Bad Characters':
        bad_chars[row['debut year']-1935] += 1
    elif row['alignment'] == 'Neutral Characters':
        neutral_chars[row['debut year']-1935] += 1
    elif row['alignment'] == 'Reformed Criminals':
        reformed_criminals[row['debut year']-1935] += 1
        
plt.figure(figsize=(15,5))
plt.bar(x_range,good_chars, label = "hero")
plt.bar(x_range,bad_chars,bottom = good_chars, label = "villain")

plt.title("Character alignment by year created", fontsize = 20)
plt.ylabel("Number of characters")
plt.xlabel("Year")

plt.legend()
```

From this graph, we can see that there is an explosion of new characters after the 80s. Furthermore, there are two a huge explosions of new characters: one happens around the 90s, and the other happens around the 2010. The most notable event that happens around the 90s is the Cold War, and so we can temporary guess that the Cold War is one of the reasons which affect the number of new comic characters. For the year around 2010, I can't guess the reason why, but one possible reason that I can think of is the 2008 financial crisis.

<hr>

## Difference between the number of heroes and villains created each year


```python
plt.figure(figsize=(15,7))
differences = [good_chars[i] - bad_chars[i] for i in range(len(good_chars))]
plt.bar(x_range,differences)
plt.title("Difference in the number of heroes created compared to villains by tear", fontsize = 20)
plt.ylabel("Number of characters")
plt.xlabel("Year")
```

This graph shows the difference between the number of heroes created and villains created each year. The values are taken by subtracting the number of villains from the number of heroes.

Here, we can see a very interesting trend. In the beginning, more heroes are created each year. This trend peaks around the year 1990, when there are roughly 100 more heroes created than villains. Furthermore, the peaks stand out completely from the rest of the years. However, this trend firmly reversed the years after 1990, and ever since 1990, there are more villains created than heroes.

<hr>

## Characters mortality rate
Now, we will see how the characters were killed off over time.

```python

# Now, let's look at the character debuts based on alignment
# 
print(dc_chars['alignment'].unique())
alive_chars = [0] * (2013-1935+1)
death_chars = [0] * (2013-1935+1)
x_range= np.arange(1935,2014)
for index, row in dc_chars.iterrows():
    if row['alive'] == 'Living Characters':
        alive_chars[row['debut year']-1935] += 1
    elif row['alive'] == 'Deceased Characters':
        death_chars[row['debut year']-1935] += 1
        
plt.figure(figsize=(15,5))
plt.bar(x_range,alive_chars, label = "alive")
plt.bar(x_range,death_chars,bottom = alive_chars, label = "deceased")
plt.title("Character mortaility by years created", fontsize = 20)
plt.ylabel("Number of characters")
plt.xlabel("Year")
plt.legend()

```

From the graph, we can see the obvious that most characters are alive. The number of overall characters who got killed off seems to be quite reasonable, and of the same proportion over the years. There is virtual no year where there are significantly more characters killed than normal. It makes sense, since this data does not reflect the year that the character was killed, but rather whether the character created that year is alive or not. 

<hr>

## Characters deceased by alignment
Let's look at only the deceased characters. 

```python
heroes_dead = [0] * (2013-1935+1)
villains_dead = [0] * (2013-1935+1)
x_range = np.arange(1935,2014)
for index, row in dc_chars.iterrows():
    if row['alive'] == 'Deceased Characters':
        if row['alignment'] == 'Good Characters': 
            heroes_dead[row['debut year']-1935] += 1
        elif row['alignment'] == 'Bad Characters':
            villains_dead[row['debut year']-1935] += 1
dead_differences = [heroes_dead[i] - villains_dead[i] for i in range(len(heroes_dead))]         
plt.figure(figsize=(15,5))
plt.title("Number of heroes killed compared to number of villains killed", fontsize = 20)
plt.ylabel("Number of characters")
plt.xlabel("Year")
plt.bar(x_range,dead_differences)
```

This graph shows the difference between the number of heroes killed and villains killed by year created. The values are taken by subtracting the number of villains from the number of heroes.

From this graph, we can see that the recently created heroes are more likely to be killed than the recently created villains. Furthermore, this trend has been very notable since after 1985s. Moreover, from the previous graphs, we also know that there are more villains created in recent years compared to heroes. This essentially means that there are more villains created and yet the villains are also less likely to be killed. 

Among the years, there are quite a few years where the number of heroes and villains debut that year which were killed were roughly the same. This trend was more prominent for characters created before 1980s. Looking at the previous graph of characters killed, this is likely because either no character was killed, or there were too few characters killed that the difference is small.

## The difference in sex among the characters created each year

```python
#What about the differences in sex?
dc_chars['sex'].value_counts()
male = [0] * (2013-1935+1)
female = [0] * (2013-1935+1)
for index, row in dc_chars.iterrows():
    if row['sex'] == 'Male Characters':
        male[row['debut year']-1935] += 1
    elif row['sex'] == 'Female Characters':
        female[row['debut year']-1935] += 1
        
x_range = np.arange(1935,2014)
plt.figure(figsize=(15,7))
sex_diffs = [male[i]-female[i] for i in range(len(male))]
plt.title("Number of males compared to females", fontsize = 20)
plt.ylabel("Number of characters")
plt.xlabel("Year")
plt.bar(x_range,sex_diffs)

def div(a,b):
    return a/b if b else 0
plt.figure(figsize=(15,7))
sex_diffs = [div(male[i],female[i]) for i in range(len(male))]
# sex_diffs = [(male[i]-female[i])/(male[i]+female[i]) for i in range(len(male))]
plt.title("Ratio between the number of male characters and female characters created by year", fontsize = 20)
plt.ylabel("Ratio")
plt.xlabel("Year")
plt.bar(x_range,sex_diffs)
```

The first graph shows the difference between the numbers of male and female characters, calculated by subtracting the number of female characters from male characters.
The second graph is created by taking the ratio between the number of male characters and female characters. 

Looking at the graphs, it is obvious that there are many more males characters than female characters, and this trend remains the same over the years. In other words, the majority of DC characters are still males. The difference is also quite big as well. 

If we only look at the difference between the number of male and female characters, then the difference between males and female characters do not seem to decrease at all. There are certain periods where the difference gets smaller, but then it got bigger again. 

However, if we look at the ratio, then we can see that the ratio actually slowly decreases over time, except for a few outlier years. 

There are quite a few years where the ratio is 0, which essentially means that there is no female character created that year. Furthermore, if we look at the previous graphs, we can also see that these outlier years do not have a lot of new characters. However, the number of such years decreases over time as well, with virtually no such year after 1980, except for year 2013. But if we looking at the data of year 2013 overall, it seems that year 2013 seems to not have enough data and can be considered data entry problem. This means that there is probably more attention toward female characters after 1980.
